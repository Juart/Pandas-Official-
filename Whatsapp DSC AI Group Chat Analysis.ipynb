```python
import pandas as pd
import matplotlib.pyplot  as plt
%matplotlib inline
```

```python
# Reading the text file
a = open('Desktop/PYTHON/DSC/Pandas Boobie/Chat.txt', encoding =  'utf8')
b = a.read()
chat = b.splitlines()
print(chat)
```

```python
Date = []
Time = []
Sender = []
Message = []
```

```python
for i in range(0, len(chat) - 1):
    if(str(chat[i].count(':')) == '2' and str(chat[i].count('/')) < '4'):
        Date.append(str.strip(chat[i][0:10]))
        Time.append(str.strip(chat[i][12:20]))
        line = chat[i][20:]
        if(line.find(':') != -1):
            Sender.append(str.strip(line[2:line.index(':')]))
            Message.append(str.strip(line[line.index(':')+1:]))
        else:
            Sender.append('?')
            Message.append('?')
```

```python
d = {'Date': Date, 'Time': Time, 'Sender': Sender, 'Message': Message}
df = pd.DataFrame(data = d)
```
